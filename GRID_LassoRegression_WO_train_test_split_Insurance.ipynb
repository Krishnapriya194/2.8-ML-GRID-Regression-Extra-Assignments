```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset=pd.read_csv('insurance_pre.csv')
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
indep=dataset[['age','bmi','children','sex_male','smoker_yes']]
indep
```

```python
dep=dataset[['charges']]
dep
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
param_grid={'alpha':[1.0],'fit_intercept':['bool',True],'precompute':['bool','array_like of shape',False],'copy_X':['bool',True],'max_iter':['int',1000]}
grid=GridSearchCV(Lasso(),param_grid,refit=True,verbose=3,n_jobs=-1)
grid.fit(indep,dep)
```

```python
re=grid.cv_results_
print("The R_score value for the best parameter{}:".format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(re)
```

```python
table
```

```python
age_input=float(input("Age:"))
bmi_input=int(input("BMI:"))
children_input=float(input("Children:"))
sex_male_input=int(input("Sex Male 0 or 1:"))
smoker_yes_input=int(input("Smoker yes 0 or 1:"))
```

```python
Future_Prediction=grid.predict([[age_input,bmi_input,children_input,sex_male_input,smoker_yes_input]])
print("Future_prediction={}".format(Future_Prediction))
```
